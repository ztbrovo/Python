## Ex5-1. 圧縮行格納方式の変種

疎行列（$0$ を多く含む行列）を表現する方法として、
$0$ でない要素のリストと、各行の $0$ でない列のインデックスのリストを用いるものがあります。
たとえば、$$\left(
\begin{array}{cccc} 1 & 2 & 3 & 0 \\ 0 & 0 & 0 & 1 \\ 2 & 0 & 0 & 2 \\ 0 & 0 & 0 & 1 \end{array}
\right)$$ という行列は、
$0$ でない要素のリスト `[1,2,3,1,2,2,1]` と、
各行の $0$ でない列のインデックスのリスト `[0,1,2]`, `[3]`, `[0,3]`, `[3]` で表されます。
各行の $0$ でない列のインデックスのリストをまとめてリストにすると、
`[[0,1,2],[3],[0,3],[3]]` となります。
（疎行列のこの表現方法は Compressed Row Storage と呼ばれる方法に近いです。）

以下のセルでは、二つのリスト `[1,2,3,1,2,2,1]` と `[[0,1,2],[3],[0,3],[3]]` をもらって、
もとの行列をリストのリストとして返す関数 `crs2matrix` を定義してください。
`crs2matrix([1,2,3,1,2,2,1], [[0,1,2],[3],[0,3],[3]])` は、
`[[1,2,3,0],[0,0,0,1],[2,0,0,2],[0,0,0,1]]` を返します。

ただし，関数 `crs2matrix` は次の点を守って下さい．

* 正方行列を返してください。
* NumPyの配列ではなくリストのリストを返してください。

```python
##########################################################
##  <[ ex5-1-crs2matrix ]> 解答セル (Answer cell)
##  このコメントの書き変えを禁ず (Never edit this comment)
##########################################################

QUESTION_EXISTS = False # 質問がある場合は True に変えて，このセル内のコメントとして質問を記述してください
                        # Change to True if you have questions, and describe them as comments in this cell

def crs2matrix(elems, colss):
    N = len(colss)
    matrix = []
    for i in range(N):
      # i行目を作る
      row = [0]* N
      for j in colss[i]:
        # i行目j列目をrowに追加する
        row[j] = elems.pop(0)
      matrix.append(row)
    return matrix
```
